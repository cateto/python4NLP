```python
text_file = "2016-10-20.txt"
```

```python
from tqdm import tqdm
from sklearn.feature_extraction.text import CountVectorizer
from konlpy.tag import Mecab
from bertopic import BERTopic
```

# 전처리

```python
documents = [line.strip() for line in open(text_file, encoding="utf-8").readlines()]
```

```python
preprocessed_documents = []

for line in tqdm(documents):
  # 빈 문자열이거나 숫자로만 이루어진 줄은 제외
  if line and not line.replace(' ', '').isdecimal():
    preprocessed_documents.append(line)
```

# 한국어 BERTopic
- 토크나이저 - Mecab
- 모델 - SBERT

```python
class CustomTokenizer:
    def __init__(self, tagger):
        self.tagger = tagger
        self.stopwords = ['으로', '에서', '지만', '때문', '에게', '습니다']
    def __call__(self, sent):
        sent = sent[:1000000]
        word_tokens = self.tagger.morphs(sent)
        tokenized = [word for word in word_tokens if len(word) > 1]
        result = []
        for token in tokenized:
          if not token in self.stopwords:
            result.append(token)
        return result
```

```python

custom_tokenizer = CustomTokenizer(Mecab())
```

```python
vectorizer = CountVectorizer(tokenizer=custom_tokenizer, max_features=3000)
```

```python
model = BERTopic(embedding_model="sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens", \
                 vectorizer_model=vectorizer,
                 nr_topics=50,
                 top_n_words=10,
                 calculate_probabilities=True)
```

```python
topics, probs = model.fit_transform(preprocessed_documents)
```

```python
model.visualize_topics()
```

```python
model.visualize_distribution(probs[0])
```

```python
for i in range(0, 50):
  print(i,'번째 토픽 :', model.get_topic(i))
```

# Coherence Score 계산
- 정량적 평가를 위한 Coherence Score 계산

```python
import pandas as pd
# 문서 가공
documents = pd.DataFrame({"Document": preprocessed_documents,
                          "ID":range(len(preprocessed_documents)),
                          "Topic":topics})
```

```python
documents_per_topic = documents.groupby(['Topic'], as_index=False).agg({'Document': ' '.join})
```

```python
cleaned_docs = model._preprocess_text(documents_per_topic.Document.values)
```

### BERTopic에서 vectorizer, analyzer 추출

```python
vectorizer = model.vectorizer_model
```

```python
analyzer = vectorizer.build_analyzer()
```

### Topic Coherence 평가에서 feature 추출

```python
import gensim.corpora as corpora
from gensim.models.coherencemodel import CoherenceModel
```

```python
words = vectorizer.get_feature_names()
tokens = [analyzer(doc) for doc in cleaned_docs]
dictionary = corpora.Dictionary(tokens)
corpus = [dictionary.doc2bow(token) for token in tokens]
topic_words = [[words for words, _ in model.get_topic(topic)] 
               for topic in range(len(set(topics))-1)]
```

### 평가

```python
# Evaluate
coherence_model = CoherenceModel(topics=topic_words, 
                                 texts=tokens, 
                                 corpus=corpus,
                                 dictionary=dictionary, 
                                 coherence='c_v')
coherence = coherence_model.get_coherence()
```
